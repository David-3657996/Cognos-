```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs FROM master
"""
df = pd.read_sql_query(QUERY,conn)

df.plot(kind='scatter', x='Age',y='WorkWeekHrs')
plt.title('relationship between Age and WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs FROM master
"""
df = pd.read_sql_query(QUERY,conn)

plt.scatter(x=df['WorkWeekHrs'], y=df['CodeRevHrs'], s=df['Age']*3)

df.plot(kind='scatter', x='WorkWeekHrs',y='CodeRevHrs',s='Age')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('the relationship among WorkWeekHrs, CodeRevHrs, and Age')
           
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT DatabaseDesireNextYear, count(Respondent) as total
FROM DatabaseDesireNextYear 
group by DatabaseDesireNextYear
order by count(Respondent) desc
limit 5
"""
df = pd.read_sql_query(QUERY,conn)
df
```

```python
plt.pie(df['total'],labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%')

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs,age
FROM master
where 30<=Age and Age<=35

"""
df = pd.read_sql_query(QUERY,conn)
df
```

```python
median_hours = df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()
median_hours
```

```python
median_hours = df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

# Create a stacked bar chart
median_hours.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')

# Display the chart
plt.show()
```

```python
# Based on the Line Chart of Age and median ConvertedComp for the age group 25-30, what kind of correlation do you see between Age and median ConvertedComp? 
QUERY = """
SELECT ConvertedComp,age
FROM master
where 25<=Age and Age<=30

"""
df = pd.read_sql_query(QUERY,conn)
median_hours = df.groupby('Age')[['ConvertedComp']].median()
df_new = median_hours.reset_index(inplace=False)
df_new
df_new.plot(kind='scatter', x='Age',y='ConvertedComp')
plt.title('The relationship between Age and median ConvertedComp for the age group 25-30')
```

```python
import numpy as np
QUERY = """
SELECT ConvertedComp,age
FROM master
where 22<=Age and Age<=60

"""
df = pd.read_sql_query(QUERY,conn)
median_hours = df.groupby('Age')[['ConvertedComp']].median()
df_new = median_hours.reset_index(inplace=False)
df_new
df_new.plot(kind='scatter', x='Age',y='ConvertedComp')

z = np.polyfit(df_new['Age'], df_new['ConvertedComp'], 1)
p = np.poly1d(z)
plt.plot(df_new['Age'], p(df_new['Age']), "r--")
plt.title('The relationship between Age and median ConvertedComp for the age group 25-60')
```
